Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
# Answer
Scatter Plot:

Use: Scatter plots are used to visualize the relationship between two numerical variables. They are excellent for showing the distribution and correlation between data points.
Function: sns.scatterplot()
Line Plot:

Use: Line plots are suitable for showing trends or changes in data over time. They are often used with time series data or when you want to connect data points to see patterns.
Function: sns.lineplot()
Bar Plot:

Use: Bar plots are used to display and compare the values of categorical variables. They are helpful for showing how different categories relate to a numerical value.
Function: sns.barplot() or sns.countplot() (for counting occurrences)
Histogram:

Use: Histograms are used to visualize the distribution of a single numerical variable. They divide the data into bins and display the frequency or density of data points in each bin.
Function: sns.histplot()
Box Plot (Box-and-Whisker Plot):

Use: Box plots are used to display the distribution and variability of data, especially for comparing multiple categories or groups. They show the median, quartiles, and any potential outliers.
Function: sns.boxplot()
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
# Answer
import seaborn as sns
import matplotlib.pyplot as plt

fmri_data = sns.load_dataset("fmri")

plt.figure(figsize=(10, 6))  

sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Line Plot of FMRI Data for Different Events and Regions")

plt.legend(title="Event", loc="upper right")

plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
# Answer
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns

sns.boxplot(x='pclass', y='age', data=titanic_data, ax=axes[0])
sns.boxplot(x='pclass', y='fare', data=titanic_data, ax=axes[1])

axes[0].set_title('Age Distribution by Pclass')
axes[1].set_title('Fare Distribution by Pclass')

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
# Answer
import seaborn as sns
import matplotlib.pyplot as plt

diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue for the 'cut' column
plt.figure(figsize=(10, 6))  

# Use Seaborn's histplot function with hue
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True, multiple='stack', palette='Set1')

plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices by Cut")

plt.legend(title="Cut", loc="upper right")

plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
# Answer
import seaborn as sns
import matplotlib.pyplot as plt

iris_data = sns.load_dataset("iris")

sns.set(style="ticks")  
sns.pairplot(iris_data, hue="species", markers=["o", "s", "D"], palette="Set1")

plt.show()
```
